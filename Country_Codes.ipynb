### Parsing Country Codes and Population Data from an HTML Page into a DataFrame

Libraries: requests, BeautifulSoup, numpy, pandas

```python
import requests
from bs4 import BeautifulSoup
page = requests.get("https://countrycode.org/")
```

```python
# Status Code 200 means that the request is successful
page.status_code
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(list(content.children)[9].find_all('td'))
```

```python
extract_content = list(content.children)[9].find_all('td')
```

```python
print(extract_content)
```

```python
country_codes = [item.get_text() for item in extract_content]
```

```python
print(country_codes)
```

```python
import pandas as pd
import numpy as np
```

```python
np_array = np.array(country_codes)
reshaped_array = np_array.reshape(-1,3)
```

```python
print(reshaped_array)
```

```python
pd.DataFrame(reshaped_array, columns={'Country', 'Code', 'ISO'})
```

```python
print(data)
```

```python
len(data)/6
```

```python
countrystats = pd.DataFrame(country_data, columns = {'Country', 'Code', 'ISO', 'Population', 'Area', 'GDP'})
```
