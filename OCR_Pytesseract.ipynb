```python
import pytesseract
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('Sunscreen.jpg')
```

```python
plt.imshow(img)
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
img_string = pytesseract.image_to_string(img)
```

```python
print(img_string)
```

```python
imgbox = pytesseract.image_to_boxes(img)
```

```python
print(imgbox)
```

```python
img_h, img_l, _ = img.shape
```

```python
for boxes in imgbox.splitlines():
    boxes = boxes.split(' ')
    x,y,z,h = int(boxes[1]), int(boxes[2]), int(boxes[3]), int(boxes[4])
    cv.rectangle(img, (x,img_h-y), (z,img_h-h), (0,0,255), 3)
```
